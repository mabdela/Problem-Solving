```python
class Card:
    """
    Here: 
    Club = 0, Diamond = 1, Heart = 2, Spade = 3
    Ace = 1, Jack = 11, Queen = 12, King = 13
    """
    suits = ["Club", "Diamond", "Heart", "Spade"]
    
    def __init__(self, s, v):
        self.rank = v
        self.s = s
    
    def getSuit(self):
        return self.s
    
    def getRank(self):
        return self.rank
    
    def toString(self):
        value = str(self.rank)
        
        if self.rank==1:
            value = "Ace"
        elif self.rank ==11:
            value = "Jack"
        elif self.rank == 12:
            value = "Queen"
        elif self.rank == 13:
            value = "King"
            
        ans = value + " of " + self.suits[self.s]
        return ans          
                
```

```python
import random 

class Deck:
    def __init__(self):
        self.cards = []
        for i in range(1, 14):
            for j in range(0, 4):
                card = Card(j, i)
                self.cards.append(card)

    def printCards(self):
        for card in self.cards:
            print(card.toString())
    
    def shuffle(self):
        
        for i in range(len(self.cards)):
            index = random.sample(range(0, len(self.cards)), 2)
            
            temp = self.cards[index[0]]
            self.cards[index[0]] = self.cards[index[1]]
            self.cards[index[1]] = temp
        
```

```python
obj = Deck()
```

```python
obj.printCards()
```

```python
obj.shuffle()
```

```python
obj.printCards()
```
