```python
"""
Given a list and a target, the function find and return indeces of pairs adding upto the target.

"""
def findPairs(arr, summ):
    
    dataStruct = {}
    
    for i in range(len(arr)):
        if(not arr[i] in dataStruct):
            dataStruct[arr[i]] = [i]
        else:
            dataStruct[arr[i]].append(i)
    
    ans =[]    
    for i in dataStruct:
        if(summ-i in dataStruct):
            for j in dataStruct[i]:
                for k in dataStruct[summ-i]:
                    if(j!=k):
                        ans.append([min(j, k), max(j, k)])
    
    unique = []
    for i in ans:
        if(i not in unique):
            unique.append(i)
    
    return sorted(unique)            
        
```

```python
"""
Given a list and a target, the function find pairs adding upto the target.
"""
def findUniquePairs(arr, summ):
    arr.sort()
    res=[]
    left, right = 0, len(arr)-1
    l, r = 0, len(arr)-1
    while(right>left):
        if(l<left and arr[left]==arr[left-1]):
            left+=1
            continue
        if(r>right and arr[right]==arr[right+1]):
            right-=1
            continue
        
        if(arr[left]+arr[right]>summ):
            right-=1
        elif(arr[left]+arr[right]<summ):
            left+=1
        else:
            res.append([arr[left], arr[right]])
            left+=1
            right-=1
    return res
            
```

```python
"""
Given an integer, it returns the reversed form. 
"""
class Solution:
    def reverse(self, x: int) -> int:
        res = 0
        neg = False
        limit = 2147483648;


        if (x<0):
            res = str(x)[1:]
            neg = True 
        else :
            res = str(x)
        res = [char for char in res]
        res.reverse()

        if (neg):
            final = int("-"+"".join(res))
        else:
            final = int("".join(res))
        if (abs(final) > limit):
            return 0
        else:
            return final
```

```python
# check if a number is a palindrome

class Solution:
    
    def checker(self, item):
        last = len(item)-1
        start =0
        while(last>start):
            if(item[last]!=item[start]):
                return False
            last-=1
            start+=1
        return True
    
    def isPalindrome(self, x: int) -> bool:
        string = str(x)
        return self.checker(string)
```

```python
"""
Find 3 sum () e.g. Given array nums = [-1, 0, 1, 2, -1, -4],A solution set is:[[-1, 0, 1],[-1, -1, 2]].
"""

class Solution:
    
    def helperFunc(self, arr, total, res, left, right):
        l, r = left, right
        
        while(left<right):
            if(l<left and arr[left-1]==arr[left]):
                left+=1
                continue
            elif(r>right and arr[right+1]==arr[right]):
                right-=1
                continue
            
            if(arr[left]+arr[right]>total):
                right-=1
            elif(arr[left]+arr[right]<total):
                left+=1
            else:
                res.append([-total, arr[left], arr[right]])
                left+=1
                right-=1
        return 
    
    def threeSum(self, nums):
        nums.sort()
        res =[]
        right = len(nums)-1
        
        for i, d in enumerate(nums):
            if(i>0 and d==nums[i-1]):
                continue
            if(d>0):
                break
            
            self.helperFunc(nums, -d, res, i+1, right)
        
        
        return res
```

```python
"""
3Sum Closest e.g. Given array nums = [-1, 2, 1, -4], and target = 1.

The sum that is closest to the target is 2. (-1 + 2 + 1 = 2).
"""


class Solution:
    def helperFunc(self, arr, target, left, right):
        l, r = left, right
        margin = 2**10
        
        while(left<right):
            if(l<left and arr[left-1]==arr[left]):
                left+=1
                continue
            elif(r>right and arr[right+1]==arr[right]):
                right-=1
                continue
                
            temp2 = arr[left]+arr[right]-target
            if(abs(temp2)<abs(margin)):
                margin = temp2

            temp3 = arr[left]+arr[right]   
            if(temp3<target):
                left+=1
            elif(temp3>target):
                right-=1
            else:
                left+=1
                right-=1
                
        return margin
    
    def threeSumClosest(self, nums, target):
        nums.sort()
        right = len(nums)-1
        closest=nums[0]+nums[1]+nums[2]
        margin = closest-target
        for i, d in enumerate(nums):
            if(i>0 and d==nums[i-1]):
                continue
            
            temp =self.helperFunc(nums, -d+target, i+1, right)
            if(abs(temp)<abs(margin)):
                closest = temp+target
                margin = temp
                
        return closest
```

```python
"""
Divide two integers without using division operator
"""
import math
class Solution:
    def divide(self, dividend: int, divisor: int) -> int:
        result = 0
        # sign = 1 (+) -1(-)
        sign = 1
        if (dividend < 0) != (divisor < 0):
            sign = -1
        x = abs(dividend)
        y = abs(divisor)
        
        mask = 1
        # binary operator
        while (y <= x):
            y <<= 1
            mask <<= 1
            
        while (mask > 1):
            y >>= 1
            mask >>= 1
            if (x >= y):
                x -= y
                result |= mask

        if sign >= 0:
            if result >= 2**31-1:
                return 2**31-1
            return int(result) * sign
        else: 
            return math.ceil(result) * sign
```

```python
"""
Need to further explain it later.
"""
class Solution:
    def nextPermutation(self, nums) -> None:
        # Nothing to swap
        if len(nums) < 2:
            return
        
        prev = float('-inf')
        for i, cur in enumerate(reversed(nums)):
            # From the right, find the first non-increasing number
            idx = len(nums) - 1 - i
            if cur >= prev:
                prev = cur
            else:
                # Once found, searching left to right, find the smallest element
                # that is still greater than cur
                minGreater = (float('inf'), None)
                for j, swp in enumerate(nums[idx:]):
                    if swp > cur:
                        minGreater = min(minGreater, (swp, idx+j), key=lambda x: x[0])
                # Swap cur and smallest greater
                nums[minGreater[1]], nums[idx] = nums[idx], nums[minGreater[1]]
                # Sort all elements after the first swap index
                nums[idx+1:] = sorted(nums[idx+1:])
                return
            
        # Otherwise, list is already in decending order, reverse it 
        for i in range(len(nums) // 2):
            nums[i], nums[len(nums)-1-i] = nums[len(nums)-1-i], nums[i] 
        return
```

```python
"""
Find the greatest common divisor for numbers in a given list.
"""
def findgcd(arr):
    
    def gcd(num1, num2):
        
        while (num2):
            
            temp = num1
            num1 = num2
            num2 = temp%num1
        return num1 
    
    ans = arr[0]
    
    for i in arr[1:]:
        ans = gcd(ans, i)
    
    return ans
```
